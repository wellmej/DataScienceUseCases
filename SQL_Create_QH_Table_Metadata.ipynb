```python
sqlDrop = 'IF Object_id([EDExtract].[dbo].[' + table_name + ']) is not null DROP TABLE [EDExtract].[dbo].[' + table_name + '];'
sqlGo = 'GO'
sqlBegin = 'CREATE TABLE '
sqlTable = '[EDExtract].[dbo].[' + table_name + ']'
sqlNULL = 'NULL'
sqlNOTNULL = 'NOT NULL'
sqlDefFileFormatId = '@DefFileFormatId'
sqlDefFileRowLayoutId = '@DefFileRowLayoutId'
sqlZero = '0'
sqlComma = ','
sqlOpenBracket = '['
sqlOpenParen = '('
sqlCloseBracket = ']'
sqlCloseParen = ')'
sqlSpace = ' '
sqlVarChar = '[varchar]'

sqlEnd = ')'

sqlConstraint = ' CONSTRAINT [PK_' + table_name + '] PRIMARY KEY CLUSTERED'
sqlCons1 = '[RowId] ASC' + sqlComma
sqlCons2 = '[FileId] ASC' + sqlComma
sqlCons3 = '[ExecutionId] ASC'
sqlWith = sqlCloseParen + 'WITH (PAD_INDEX = OFF, STATISTICS_NORECOMPUTE = OFF, IGNORE_DUP_KEY = OFF, ALLOW_ROW_LOCKS = ON, ALLOW_PAGE_LOCKS = ON) ON [PRIMARY]'
sqlStorage = sqlCloseParen + ' ON [PRIMARY]'
i = 0
```

```python
for ind, row in df_database_info.iterrows():
    if row['Type'] == 'v':
        df_database_info.loc[ind, "SQL_Create_Tbl"] = sqlOpenBracket + row['Column_Header'] + sqlCloseBracket + sqlSpace  + sqlVarChar + sqlOpenParen +  str(row['Length_of_Field']) + sqlCloseParen + sqlSpace + sqlNULL + sqlComma
```

```python
#f = open(out_file, 'w', encoding="latin1")
f = open(out_file, 'w')
f.write(sqlDrop + '\n')
f.write(sqlGo + '\n')
f.write(sqlBegin + sqlTable + '\n')
f.write(sqlOpenParen + '\n')
for ind, row in df_database_info.iterrows():
    if i > len(df_database_info):
       break
    else:
#       f = open(str(i)+'.txt', 'w')
       f.write(row ['SQL_Create_Tbl'] + '\n')
       print (row['SQL_Create_Tbl']) 
       i+=1
#
f.write(sqlConstraint + '\n')
f.write(sqlOpenParen + '\n')
f.write(sqlCons1 + '\n')
f.write(sqlCons2 + '\n')
f.write(sqlCons3+ '\n')
f.write(sqlWith + '\n')
f.write(sqlStorage + '\n')
#
f.close()
```

```python
#------------------------------------------------
# Set full file path names
#------------------------------------------------
folder_name = '/content/drive/My Drive/Colab Notebooks/'
file_name = 'Database_Table_Definition_Info.txt'
csv_in_file = folder_name + file_name

out_file_name = 'Database Table Definition Info Output.sql'
out_file = folder_name + out_file_name
```

```python
#------------------------------------------------
# Read .CSV file into a dataframe
#------------------------------------------------
df_database_info = pd.read_csv(csv_in_file)
df_database_info.head(10)
```

```python
sqlBegin = 'insert into DICE.dbo.DAT_FileFormat values ('
sqlNULL = 'NULL'
sqlDefFileFormatId = '@DefFileFormatId'
sqlDefFileRowLayoutId = '@DefFileRowLayoutId'
sqlZero = '0'
sqlFieldStart = 0
sqlEnd = ')'
i = 0
```

```python
for ind, row in df_database_info.iterrows():
    df_database_info.loc[ind, "New_SQL"] = sqlBegin + sqlDefFileFormatId + ',' + "'" + row['Column_Header'] + "'" + ',' + str(row['Length_of_Field']) + ',' + str(row['Column_Order']) + ',' + str(sqlFieldStart) + ',' + "'" + row['Column_Header'] + "'" + ',' + sqlZero + ',' + sqlDefFileRowLayoutId + ',' + sqlZero + ',' + sqlNULL + sqlEnd
    sqlFieldStart = sqlFieldStart + row['Length_of_Field']
```

```python
#----------------------------------------------------
# Write out DICE.dbo.DAT_FileFormat values          
#----------------------------------------------------
f = open(out_file, 'w')
for ind, row in df_database_info.iterrows():
    if i > len(df_database_info):
       break
    else:
#       f = open(str(i)+'.txt', 'w')
       f.write(row['New_SQL'] + '\n')
       print (row['New_SQL']) 
       i+=1
#
#
f.close()
```
